```python
import folium
import random

# Coordenadas de los hospitales
latitudes_hospitales = [-12.0411954, -12.0216057, 25.797809, 38.176505, -12.05963535, -12.0782058, -11.96637165, -12.055924, -11.9141137, -12.07442605]
longitudes_hospitales = [-76.99263863265921, -77.05509942443221, -100.5616261, -0.7983121, -77.02236228077756, -77.0399864964742, -77.05641452313083, -77.01569082772036, -77.03751289728125, -76.95597427784205]

# Función para generar ubicaciones aleatorias dentro del área de los hospitales
def generar_ubicacion():
    # Límites de las coordenadas de los hospitales
    latitud_min = -11.894451
    latitud_max = -12.097977
    longitud_min = -76.967502
    longitud_max = -77.081000
    
    # Generar latitud y longitud aleatorias dentro del área de los hospitales
    latitud = random.uniform(latitud_min, latitud_max)
    longitud = random.uniform(longitud_min, longitud_max)
    
    return latitud, longitud

# Función para generar severidad aleatoria entre 1 y 4
def generar_severidad():
    return random.randint(1, 4)

# Crear mapa centrado en la primera ubicación de los hospitales
mapa = folium.Map(location=[latitudes_hospitales[0], longitudes_hospitales[0]], zoom_start=12)

# Iterar sobre las listas de latitudes y longitudes de los hospitales y agregar marcadores al mapa
for lat, long in zip(latitudes_hospitales, longitudes_hospitales):
    folium.Marker(location=[lat, long], popup='Hospital', icon=folium.Icon(color='blue', icon='hospital')).add_to(mapa)

# Generar 1000 ubicaciones aleatorias con severidades y agregar marcadores al mapa
for i in range(1000):
    latitud, longitud = generar_ubicacion()
    severidad = generar_severidad()
    
    if severidad == 1:
        color = 'green'
    elif severidad == 2:
        color = 'yellow'
    elif severidad == 3:
        color = 'orange'
    else:
        color = 'red'
    
    # Agregar marcador de paciente al mapa con el color correspondiente
    folium.Marker(location=[latitud, longitud], icon=folium.Icon(color=color, icon_size=(10, 10)), popup=f'Severidad: {severidad}').add_to(mapa)

# Guardar el mapa como un archivo HTML
mapa.save('mapa_hospitales_y_severidades.html')
```
